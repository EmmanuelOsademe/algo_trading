```python
data =  yf.download(tickers=['AAPL'], start= '2024-01-01', end= '2024-06-30',group_by= 'Ticker')
data.info()
```

```python
df = data
open_data = df["AAPL"]["Open"]
```

```python
plt.plot(open_data)
plt.ylabel("Opening Prices ($)")
plt.title("Apple Stock Openinig Prices")
plt.show()
```

```python
plt.plot(open_data, label="Opening Price", color="Red")
plt.plot(df["AAPL"]["Close"], label="Closing Price", color="Blue")
plt.ylabel("Prices ($)")
plt.title("Apple Stock Openinig and Closing Prices")
plt.legend()
plt.show()
```

```python
plt.plot(df["AAPL"]["High"], label="High Price", color="Green", linestyle="dashed", linewidth=2)
plt.plot(df["AAPL"]["Low"], label="Low Price", color="Red", linestyle="dotted", linewidth=2)
plt.xlabel("Time")
plt.ylabel("Prices ($)")
plt.title("Apple Open and Close Price Comparison")
plt.legend()
plt.show()
```

```python
plt.plot(df["AAPL"]["High"], label="High", color="Green", marker="d", markersize=3)
plt.plot(df["AAPL"]["Low"], label="Low", color="Red", marker="o", markersize=3)
plt.xlabel("Time")
plt.ylabel("Prices ($)")
plt.title("Apple Open and Close Price Comparison")
plt.legend(loc="upper center")
plt.ylim(160,230)
plt.grid()
plt.show()
```
