```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
data = yf.download(tickers="AAPL", start="2024-01-01", end="2024-12-31")
data.head()
```

```python
def fetch_historical_data(tickers, start_date, end_date):
    data = yf.download(tickers=tickers, start=start_date, end=end_date)
    return data

tickers = ["AAPL", "MSFT", "GOOGL", "AMZN", "META", "TSLA"]
start_date = '2024-05-01'
end_date = '2024-06-20'

df = fetch_historical_data(tickers, start_date, end_date)
df.head()
```

```python
# Multi-index-list

m_index = [("AAPL", 1), ("TSLA", 2), ("MSFT", 3), ("META", 4), ("IOCL", 5)]
m_series = pd.Series([100, 200, 300, 400, 500], index = m_index)
print(m_series)
```

```python
print(m_series[("AAPL", 1)])
```

```python
m_index = [("AAPL", 1), ("TSLA", 2), ("MSFT", 3), ("META", 4), ("IOCL", 5)]
m_df = pd.DataFrame([100, 200, 300, 400, 500], index = m_index)
print(m_df)
print(m_df.iloc[0, 0])
```

```python
t_index = pd.MultiIndex.from_tuples(m_index)
print(t_index.levels)
print(t_index.levels[0])
```

```python
p_index = [["TSLA", "MSFT", "META"], [100, 200]]
p_index = pd.MultiIndex.from_product(p_index)
print(p_index.levels)
print(p_index.levels[0])
```

```python
m_series = pd.Series([100, 200, 300, 400, 500], index=t_index)
print(m_series)
print(m_series["AAPL"])
print(m_series.iloc[0])
print(m_series.unstack())
print(m_series.unstack().stack())
```

```python
m_df = pd.DataFrame([100, 200, 300, 400, 500], index=t_index)
print(m_df)
print(m_df.loc["AAPL", 0])
print(m_df.unstack())
print(m_df.unstack().stack())
```
